```python
def plot_group(x):
    
    ## The function plot histogram, box plot, 
    ## violin plot, and density distribution plots for each 
    ## feature name passed as the argument
    
    fig = plt.figure(figsize=(16, 8))
    plt.subplots_adjust(hspace = 0.6)
    sns.set_palette('Pastel1')
    
    plt.subplot(221)
    ax1 = sns.distplot(train_set[x], color = 'purple')
    ax1.axvline(
        np.mean(train_set[x]), color="purple", linestyle="--"
    )  # Add mean to the histogram
    ax1.axvline(
        np.median(train_set[x]), color="black", linestyle="-"
    )  # Add median to the histogram
    plt.title(f'{x} Density Distribution', fontweight='bold' , fontsize = 15, fontfamily='Serif')
    
    plt.subplot(222)
    ax2 = sns.violinplot(x = train_set[x], palette = 'Accent')
    plt.title(f'{x} Violinplot', fontweight='bold' , fontsize = 15, fontfamily='Serif')
    
    plt.subplot(223)
    ax3 = sns.boxplot(x=train_set[x], palette = 'cool', width=0.7, linewidth=0.6, showmeans=True)
    plt.title(f'{x} Boxplot', fontweight='bold' , fontsize = 15, fontfamily='Serif')
    
    plt.subplot(224)
    ax4 = sns.kdeplot(train_set[x],  fill = True)
    plt.title(f'{x} Cumulative Density Distribution' , fontweight='bold' , fontsize = 15, fontfamily='Serif')
    
    plt.show()
```

```python
def bionary_plots(X):
    plt.figure(figsize=(6,4));   
    ax=sns.countplot(x = X , data=train_set ,  palette='rocket_r')
    for p in ax.patches:
         #Place the percentage of each label at the top of the plot
        percentage = '{:.1f}%'.format(100 * p.get_height()/len(train_set[X]))    
       
        x = p.get_x() + p.get_width() / 2
        y = p.get_height() + 15
        plt.annotate(percentage, (x, y), ha='center', color='black', fontweight='bold')
        plt.xticks(color='black', fontsize='medium', ticks=[0,1], labels=['No', 'Yes'] , fontweight='bold')
    plt.tight_layout()
    plt.title(f'{X} Distribution by Percentage', color='black' , fontweight='bold' , fontsize = 15)
    plt.xlabel(f'{X} ' , fontweight='bold' , fontsize = 15)
    plt.ylabel('Count' , fontweight='bold' , fontsize = 15)
    plt.show()
```

```python
## Create a function that returns a Pie chart and a Bar Graph for the categorical variables

def cat_plots(X):
    plt.figure(figsize=(15 , 4))
    plt.subplot(1 , 2 , 1)
    
    # Draw a Pie Chart on first subplot
    
    labels = train_set[X].value_counts().index
    colors = sns.color_palette('Dark2')
    plt.pie(train_set[X].value_counts() , labels = labels , textprops = {'color': 'Green','fontsize':16} ,  
            colors = colors , autopct = '%0.0f%%' , shadow = 'True')
    plt.title(f'{X} Pie Chart' , fontsize = 15 , fontweight='bold')
    plt.subplot(1 , 2, 2)
    
    ## Draw a Bar Graph on second subplot
    
    sns.countplot(x=X, hue="price_range", data=train_set , palette='rocket_r')
    plt.title(f'{X} Bar Chart' , fontsize = 15 , fontweight='bold')
    plt.xlabel(f'{X}' , fontsize = 10 , fontweight='bold')
    plt.ylabel('Count' , fontsize = 10 , fontweight='bold')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5) , shadow=True , labels =['Low Cost (0)', 'Medium Cost (1)' , 'High Cost (2)' , 'Very High Cost (3)'])
    plt.show()
```

## **Data Normalization**

### *MinMaxScaler*

```python
cols = ['battery_power' , 'clock_speed' , 'fc' , 'int_memory' , 'mobile_wt' ,
        'n_cores' , 'pc' , 'px_height' , 'px_width' , 'ram' , 'sc_h' , 'sc_w' , 'talk_time']

normz = MinMaxScaler()

train_set[cols] = normz.fit_transform(train_set[cols])
train_set
```

### *StandardScaler*

## **Model Selection**

### *Split X , y*

```python
train_set = train_set.values
X = np.delete(train_set , -1 , axis=1)
y = train_set[: , [-1]]
X.shape , y.shape
```

### *Decision Tree*

```python
model = DecisionTreeClassifier(random_state=4)
```

```python
max_depth = [3 , 4, 5 , 6 , 7 , 8]
criterion = ['entropy' , 'gini']
param_distr = {'max_depth': max_depth, 'criterion': criterion}
```

#### *Use `RandomizedSearchCV`*

```python
random_search = RandomizedSearchCV(model, param_distr)
random_search.fit(X, y.ravel())

random_search.best_params_, random_search.best_score_
```

#### *Best Model*

```python
model_tree = random_search.best_estimator_
```

#### *Evaluate on Train Set*

```python
report_DTs = classification_report(y , model_tree.predict(X))
print(report_DTs)
```

```python
cf_matrix = confusion_matrix(y , model_tree.predict(X))

plt.figure(figsize = (8 , 6))
ax = sns.heatmap(cf_matrix, annot=True, fmt='', cmap='summer')

ax.set_title('Seaborn Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['predicted 0s' , 'predicted 1s' , 'predicted 2s' , 'predicted 3s'] , rotation = 90)
ax.yaxis.set_ticklabels(['Actual 0s' , 'Actual 1s', 'Actual 2s' , 'Actual 3s'], rotation = 90)

## Display the visualization of the Confusion Matrix.
plt.show()
```

### *Random Forest*

```python
model = RandomForestClassifier(random_state=5)
```

```python
max_depth = [3 , 4, 5 , 6 , 7 , 8]
n_estimators = np.linspace(50 , 300 , 50 , dtype='int')
criterion = ['entropy' , 'gini']
param_distr = {'max_depth': max_depth, 'n_estimators':n_estimators  , 'criterion': criterion}
```

#### *Use `RandomizedSearchCV`*

```python
random_search = RandomizedSearchCV(model, param_distr)
random_search.fit(X, y.ravel())

random_search.best_params_, random_search.best_score_
```

#### *Best Model*

```python
model_rf = random_search.best_estimator_
```

#### *Evaluate on Train Set*

```python
report_rf = classification_report(y , model_rf.predict(X))
print(report_rf)
```

```python
cf_matrix = confusion_matrix(y , model_rf.predict(X))

plt.figure(figsize = (8 , 6))
ax = sns.heatmap(cf_matrix, annot=True, fmt='', cmap='summer')

ax.set_title('Seaborn Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['predicted 0s' , 'predicted 1s' , 'predicted 2s' , 'predicted 3s'] , rotation = 90)
ax.yaxis.set_ticklabels(['Actual 0s' , 'Actual 1s', 'Actual 2s' , 'Actual 3s'], rotation = 90)

## Display the visualization of the Confusion Matrix.
plt.show()
```

### *SVM*

```python
model = svm.SVC(random_state=4)
```

```python
gamma = ['scale', 'auto']
C = np.logspace(-3 , 3 , 7)
kernel = ['rbf' , 'poly' , 'sigmoid', 'linear']
param_distr = {'gamma': gamma, 'kernel':kernel  , 'C': C}
```

#### *Use `RandomizedSearchCV`*

```python
random_search = RandomizedSearchCV(model, param_distr)
random_search.fit(X, y.ravel())

random_search.best_params_, random_search.best_score_
```

#### *Best Model*

```python
model_svm = random_search.best_estimator_
```

#### *Evaluate on Train Set*

```python
report_svm = classification_report(y , model_svm.predict(X))
print(report_svm)
```

```python
cf_matrix = confusion_matrix(y , model_svm.predict(X))

plt.figure(figsize = (8 , 6))
ax = sns.heatmap(cf_matrix, annot=True, fmt='', cmap='summer')

ax.set_title('Seaborn Confusion Matrix\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['predicted 0s' , 'predicted 1s' , 'predicted 2s' , 'predicted 3s'] , rotation = 90)
ax.yaxis.set_ticklabels(['Actual 0s' , 'Actual 1s', 'Actual 2s' , 'Actual 3s'], rotation = 90)

## Display the visualization of the Confusion Matrix.
plt.show()
```
